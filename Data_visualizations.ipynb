```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv('sample_all.csv')
NoConvData = pd.read_csv('sample_all_NoConv.csv')
PosData = pd.read_csv('sample_all_Positions.csv')
```

```python
NoConvData.loc[NoConvData['TRY_Convert'] == 1, 'Points'] = 6
NoConvData['Points']
```

```python
total = []
for j in Data['PositionId'].unique():
    DataBack = Data[Data['PositionId'] == j]
    points = []
    years = DataBack['SeasonId'].unique()
    for i in years:
        temp = DataBack[DataBack['SeasonId'] == i]
        points.append(temp['Points'].sum())
    total.append(points)
fig = plt.figure(figsize=(15,10))
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.bar(years + 0.00, total[1], color = 'b', width = 0.22, label = 'Back')
ax.bar(years + 0.25, total[0], color = 'g', width = 0.22, label = 'Mid')
ax.bar(years + 0.5, total[2], color = 'r', width = 0.22, label = 'Front')
plt.legend()
plt.xlabel('Year')
plt.xticks([2019.25,2020.25,2021.25], ['2019', '2020', '2021'])
plt.ylabel('Points')
plt.title(f"Points per year per field block")
plt.savefig('PPPosition', dpi = 200)
plt.show
```

```python
total = []
years = [2019,2020,2021]
temp = Data[['SeasonId','Goal_TRY','Goal_CVOK','Goal_PGOK','Goal_FGOK']]
for col in temp.columns:
    
    if col == 'SeasonId':
        continue
    points = []
    for year in years:
        temp1 = temp[temp['SeasonId'] == year]
        temp1 = temp1[col]
        point = temp1.sum()/len(Data[Data['SeasonId'] == year])
        points.append(point)
    total.append(points)
print(total)
fig = plt.figure(figsize=(15,10))
ax = fig.add_axes([0.1,0.1,0.8,0.8])
rect1 = ax.bar([2019,2020,2021], total[0], color = 'b', width = 0.19, label = 'Tries')
rect2 = ax.bar([2019.2,2020.2,2021.2], total[1], color = 'g', width = 0.19, label = 'Conversions')
rect3 = ax.bar([2019.4,2020.4,2021.4], total[2], color = 'r', width = 0.19, label = 'Penalties')
rect4 = ax.bar([2019.6,2020.6,2021.6], total[3], color = 'y', width = 0.19, label = 'Field goals')

plt.legend()
plt.xlabel('Year')
plt.xticks([2019.33,2020.33,2021.33], ['2019', '2020', '2021'])
plt.ylabel('Percentage of goals by type by year')
plt.title(f"Breakdown of how points are being scored")

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.01*height,
                f"{'%.1f' % (height*100)}%",
                ha='center', va='bottom')

autolabel(rect1)
autolabel(rect2)
autolabel(rect3)
autolabel(rect4)

plt.savefig('Point_breakdown', dpi = 500)
plt.show
```
