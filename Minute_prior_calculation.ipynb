```python
import pandas as pd
import numpy as np
```

```python
Local_dir = "K:\\Users\KJR\\Documents\\CSVs\\DATA3001\\" #replace however you need to
Event_list = pd.read_csv(Local_dir + "2019-2021_EventData.csv")
```

```python
Events = Event_list[Event_list['Points'] > 0]
Events
```

```python
match_list = pd.read_csv(Local_dir + "2019-2021_MatchListing.csv")
```

```python
match_list['MatchId']
matches = match_list.sample(n=300)
matches = matches['MatchId']
matches
```

```python
Events = Events[Events['MatchId'].isin(matches)]
Events
```

```python
ind = Events.index
plays = []
j = 0
for i in ind:    
    time = Events['ElapsedMillisecs'].loc[i]
    half = Events['Half'].loc[i]
    pre_time = time - 60000
    match = Events['MatchId'].loc[i]
    Elist = Event_list[Event_list['MatchId'] == match]
    Elist = Elist[Elist['Half'] == half]
    small_plays = Elist[Elist['ElapsedMillisecs'] >= pre_time]
    small_plays = small_plays[small_plays['ElapsedMillisecs'] <= time]
    play = small_plays.iloc[0]
    plays.append(play.name)
```

```python
E = list(Events.index)
```

```python
difference = []

zip_object = zip(E, plays)
for E_i, plays_i in zip_object:
    difference.append(E_i-plays_i)

print(np.mean(difference))
```
