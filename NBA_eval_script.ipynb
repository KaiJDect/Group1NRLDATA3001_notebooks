```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_1samp
```

```python
nba = pd.read_csv('nba.csv')
```

```python
mu = nba.mean()
sigma = nba.std()

print(mu, sigma)
```

```python
nba['PPG diff'] = nba['PPG'] - nba['PPG'].shift(1)
nba['3PA diff'] = nba['3PA'] - nba['3PA'].shift(1)
nba = nba.fillna(0)
nba
```

```python
test = nba.drop(19)
test
```

```python
tset, pval = ttest_1samp(test['3PA diff'], 3)

print('p-values',pval)
if pval < 0.05:    # alpha value is 0.05 or 5%
   print(" we are rejecting null hypothesis that there was no significant increase in 3PA in 2018-19")
else:
  print("we are accepting null hypothesis that there was no significant increase in 3PA in 2018-19")
```

```python
tset, pval = ttest_1samp(test['PPG diff'], 3)

print('p-values',pval)
if pval < 0.05:    # alpha value is 0.05 or 5%
   print(" we are rejecting null hypothesis that there was no significant increase in PPG in 2018-19")
else:
  print("we are accepting null hypothesis that there was no significant increase in PPG in 2018-19")
```
